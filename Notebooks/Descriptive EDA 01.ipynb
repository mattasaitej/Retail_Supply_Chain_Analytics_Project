```python
# Remove incomplete final month
df_filtered = df[df['order date (DateOrders)'] < '2018-01-01']

# Monthly sales
monthly_sales = (df_filtered.groupby(df_filtered['order date (DateOrders)'].dt.to_period('M'))['Sales'].sum())
months = monthly_sales.index.astype(str)

# Rolling average
rolling_sales = (
    monthly_sales
    .rolling(window=3)
    .mean()
)

# Monthly Growth %
monthly_growth = (monthly_sales.pct_change() * 100)


fig, ax1 = plt.subplots(figsize=(15,6))
# Sales lines
ax1.plot(months,monthly_sales.values,marker='o',label='Monthly Sales')

ax1.plot(months,rolling_sales.values,marker='s',label='3-Month Rolling Avg')

ax1.set_xlabel('Month')
ax1.set_ylabel('Sales')
ax1.tick_params(axis='x', rotation=90)

# Secondary axis for growth %
ax2 = ax1.twinx()

ax2.plot(months,monthly_growth.values,marker='^',linestyle='--',label='Monthly Growth %')

ax2.set_ylabel('Growth (%)')

# Combined legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2,labels1 + labels2,loc='lower left')
plt.title('Sales Trend, Seasonality & Growth Over Time')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create temp dataframe
temp_df = df[['Product Name', 'Sales', 'Order Profit Per Order']]

# Combine same products into one
product_summary = temp_df.groupby('Product Name').agg({
    'Sales': 'sum',
    'Order Profit Per Order': 'sum'
}).reset_index()

# Calculate overall profit margin for each product
product_summary['Profit Margin'] = (
    product_summary['Order Profit Per Order'] /
    product_summary['Sales']
) * 100

product_summary.head(10)
```

```python
high_margin_low_sales = product_summary[
    (product_summary['Profit Margin'] > product_summary['Profit Margin'].median()) &
    (product_summary['Sales'] < product_summary['Sales'].median())
][['Product Name', 'Sales', 'Profit Margin']]

high_margin_low_sales = high_margin_low_sales.sort_values(
    by='Profit Margin',
    ascending=False
)

print("High Margin + Low Sales")
high_margin_low_sales.head(20)
```

```python
low_margin_high_sales = product_summary[
    (product_summary['Profit Margin'] < product_summary['Profit Margin'].median()) &
    (product_summary['Sales'] > product_summary['Sales'].median())
][['Product Name', 'Sales', 'Profit Margin']]

low_margin_high_sales = low_margin_high_sales.sort_values(
    by='Profit Margin',
    ascending=True
)

print("Low Margin + High Sales")
low_margin_high_sales.head(20)
```

```python
# Top 3 and least 3 products
top_3 = product_summary.nlargest(3, 'Sales')
least_3 = product_summary.nsmallest(3, 'Sales')

label_products = pd.concat([top_3, least_3])

plt.figure(figsize=(18,15))

# Bubble chart
plt.scatter(
    product_summary['Sales'],
    product_summary['Profit Margin'],
    s=abs(product_summary['Order Profit Per Order']) / 50,
    alpha=0.6
)

# Log scale
plt.xscale('log')

# Median lines
sales_med = product_summary['Sales'].median()
margin_med = product_summary['Profit Margin'].median()

plt.axvline(sales_med, linestyle='--')
plt.axhline(margin_med, linestyle='--')

# Better label alignment (offset text)
for _, row in label_products.iterrows():
    plt.annotate(
        row['Product Name'],
        xy=(row['Sales'], row['Profit Margin']),
        xytext=(8, 6),  # offset from bubble
        textcoords='offset points',
        fontsize=9
    )

# Axis limits
xmin = product_summary['Sales'].min()
xmax = product_summary['Sales'].max()
ymin = product_summary['Profit Margin'].min()
ymax = product_summary['Profit Margin'].max()

# Proper quadrant text placement
plt.text(
    xmin*1.2, ymax*0.9,
    'Low Sales\nHigh Profit',
    fontsize=11,color="red"
)

plt.text(
    xmax/3, ymax*0.9,
    'High Sales\nHigh Profit',
    fontsize=11,color="red"
)

plt.text(
    xmin*1.2, ymin*0.85,
    'Low Sales\nLow Profit',
    fontsize=11,color="red"
)

plt.text(
    xmax/3, ymin*0.85,
    'High Sales\nLow Profit',
    fontsize=11,color="red"
)

plt.xlabel('Total Sales (Log Scale)')
plt.ylabel('Profit Margin (%)')
plt.title('Product Performance: Sales vs Profit Margin')

plt.grid(True, alpha=0.3)
plt.tight_layout()  # fixes clipping

plt.show()
```
